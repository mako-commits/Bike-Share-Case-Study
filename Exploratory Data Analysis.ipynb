```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./cleaned_data.csv", parse_dates=['started_at', 'ended_at'])
df.head()
```

```python
# Checking conversion
data_types = df.dtypes

# Display the result
print(data_types)
```

```python
# Calculate the mean of 'numeric_column'
df['ride_duration'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60  # in minutes

# Create a new column 'day_of_week' with the day of the week
df['day_of_week'] = df['started_at'].dt.day_name()

df.head()
```

##### Averge Ride Time

```python
# Calculate average, maximum, and minimum ride times
average_ride_time = df['ride_duration'].mean()
max_ride_time = df['ride_duration'].max()
min_ride_time = df['ride_duration'].min()

print(f"Average Ride Time: {average_ride_time:.2f} minutes")
print(f"Maximum Ride Time: {max_ride_time:.2f} minutes")
print(f"Minimum Ride Time: {min_ride_time:.2f} minutes")
```

```python
# Find the row with the maximum ride duration
max_duration_row = df.loc[df['ride_duration'].idxmax()]

# Find the row with the minimum ride duration
min_duration_row = df.loc[df['ride_duration'].idxmin()]

# Display the results
print("Row with Maximum Ride Duration:")
print(max_duration_row)

print("\nRow with Minimum Ride Duration:")
print(min_duration_row)
```

```python
negative_durations_count = (df['ride_duration'] < 0).sum()

print(f"Number of rows with negative durations: {negative_durations_count}")
```
